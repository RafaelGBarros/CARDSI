```python
import pandas as pd
```

```python
datas = {'name': ['Argentina','Brasil','França','Italia','Reino Unido'],
         'Continente':['ameria','america','europa','europa','europa'],
         'Extensão':['999','999','999','999','999'],
         'CorVerd':['0','1','0','1','0'],
         }
codes = ['BR','AR','FR','IT','UK']

countries = pd.DataFrame(datas, index=codes)

has_BR = 'BR' in countries.index
has_US = 'US' in countries.index
print(has_BR)
print(has_US)

has_corVerde = 'CorVerd' in countries.index
has_corAzul = 'corAzul' in countries.index
print('\n')
print(has_corVerde)
print(has_corAzul)


has_Brasil = countries['name'].isin(['Brasil'])
print('\n')
print(has_Brasil)
```

```python
print('---  DataFrame Original  ---')
print(countries)

countries.loc['JP'] = {
    'nome': 'Japão',
    'Continente': 'Asia',
    'Extensão': 'JP',
    'CorVerd': '0'
}


print('\n')
print('---  DataFrame Com o Japão  ---')
countries.at['JP', 'name']='Japão'
print(countries)


print('\n')
print('---  DataFrame eliminando linhas  ---')
countries = countries.drop(['AR','UK'])
print(countries)
```
